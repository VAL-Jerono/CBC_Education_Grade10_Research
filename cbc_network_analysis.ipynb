## Step 3: Generate Edges — Connections via Shared Themes

Two sources are connected when they **share at least one research theme**.  
Edge weight = number of shared themes (stronger connection = thicker edge).

This is **co-theme linkage** — the standard method in bibliometric network analysis.


```python
# ── EDGE GENERATION: shared theme linkage ───────────────────────────────────
edges_data = []

for i, row_i in df_nodes.iterrows():
    for j, row_j in df_nodes.iterrows():
        if j <= i:
            continue
        shared = set(row_i["themes"]) & set(row_j["themes"])
        if shared:
            edges_data.append({
                "source":        row_i["id"],
                "target":        row_j["id"],
                "weight":        len(shared),
                "shared_themes": list(shared)
            })

df_edges = pd.DataFrame(edges_data)
print(f"✅ {len(df_edges)} edges generated")
print(f"   Min shared themes per edge : {df_edges['weight'].min()}")
print(f"   Max shared themes per edge : {df_edges['weight'].max()}")
print(f"   Mean shared themes         : {df_edges['weight'].mean():.2f}")
print()

# Top 10 strongest connections
print("Top 10 strongest source-to-source connections:")
top = df_edges.nlargest(10, "weight")[["source","target","weight","shared_themes"]]
print(top.to_string(index=False))
```

## Step 4: Build the NetworkX Graph

```python
# ── BUILD GRAPH ─────────────────────────────────────────────────────────────
G = nx.Graph()

# Add nodes with all attributes
for _, row in df_nodes.iterrows():
    G.add_node(row["id"],
               label=row["label"],
               year=row["year"],
               ntype=row["type"],
               level=row["level"],
               region=row["region"],
               themes=row["themes"],
               weight=row["weight"])

# Add weighted edges
for _, row in df_edges.iterrows():
    G.add_edge(row["source"], row["target"],
               weight=row["weight"],
               shared_themes=row["shared_themes"])

print("✅ Graph constructed")
print(f"   Nodes (vertices) : {G.number_of_nodes()}")
print(f"   Edges            : {G.number_of_edges()}")
print(f"   Graph density    : {nx.density(G):.4f}  (1.0 = fully connected)")
print(f"   Is connected     : {nx.is_connected(G)}")
avg_deg = sum(dict(G.degree()).values()) / G.number_of_nodes()
print(f"   Average degree   : {avg_deg:.2f} connections per node")
```

## Step 5: Core Network Metrics

We compute four key metrics. Each tells a different story:

| Metric | What it measures | Why it matters for funders |
|---|---|---|
| **Degree centrality** | How many other sources a node connects to | = how influential / how cited a topic is |
| **Betweenness centrality** | How often a node sits on the shortest path between others | = bridge nodes — key connectors between clusters |
| **Closeness centrality** | How close a node is to all others | = how central to the whole debate |
| **Eigenvector centrality** | Connected to well-connected nodes | = importance by association |


```python
# ── CENTRALITY METRICS ──────────────────────────────────────────────────────
degree_cent      = nx.degree_centrality(G)
betweenness_cent = nx.betweenness_centrality(G, weight='weight', normalized=True)
closeness_cent   = nx.closeness_centrality(G)
eigenvector_cent = nx.eigenvector_centrality(G, weight='weight', max_iter=1000)

# Compile into DataFrame
df_metrics = pd.DataFrame({
    "node":             list(degree_cent.keys()),
    "degree_cent":      list(degree_cent.values()),
    "betweenness_cent": list(betweenness_cent.values()),
    "closeness_cent":   list(closeness_cent.values()),
    "eigenvector_cent": list(eigenvector_cent.values()),
})
df_metrics["label"] = df_metrics["node"].map(df_nodes.set_index("id")["label"])
df_metrics["type"]  = df_metrics["node"].map(df_nodes.set_index("id")["type"])

# Composite influence score
df_metrics["influence_score"] = (
    df_metrics["degree_cent"]      * 0.30 +
    df_metrics["betweenness_cent"] * 0.30 +
    df_metrics["closeness_cent"]   * 0.20 +
    df_metrics["eigenvector_cent"] * 0.20
)

df_metrics = df_metrics.sort_values("influence_score", ascending=False).reset_index(drop=True)

print("✅ Centrality metrics computed")
print()
print("Top 10 Most Influential Nodes (by composite influence score):")
cols = ["label","type","degree_cent","betweenness_cent","closeness_cent","influence_score"]
print(df_metrics[cols].head(10).round(4).to_string(index=False))
```

## Step 6: Theme Frequency Analysis

```python
# ── THEME FREQUENCY ─────────────────────────────────────────────────────────
all_themes = []
for _, row in df_nodes.iterrows():
    all_themes.extend(row["themes"])

theme_counts = Counter(all_themes)
df_themes = pd.DataFrame(theme_counts.items(), columns=["theme","frequency"]).sort_values("frequency", ascending=False)

print("✅ Theme frequency analysis")
print()
print("Theme frequency across all 24 sources:")
print(df_themes.to_string(index=False))
```

## Step 7: Community Detection — Cluster Analysis

```python
# ── COMMUNITY DETECTION (Greedy Modularity) ─────────────────────────────────
from networkx.algorithms.community import greedy_modularity_communities

communities = list(greedy_modularity_communities(G, weight='weight'))
modularity  = nx.community.modularity(G, communities, weight='weight')

# Assign community ID to each node
node_community = {}
for cid, comm in enumerate(communities):
    for node in comm:
        node_community[node] = cid

# Label communities by their dominant type
community_labels = {}
for cid, comm in enumerate(communities):
    types = [G.nodes[n]['ntype'] for n in comm]
    dominant = Counter(types).most_common(1)[0][0]
    community_labels[cid] = f"Cluster {cid+1} ({dominant})"

print(f"✅ {len(communities)} communities detected")
print(f"   Modularity score: {modularity:.4f}  (>0.3 = meaningful structure)")
print()
for cid, comm in enumerate(communities):
    labels = [G.nodes[n]['label'] for n in comm]
    print(f"  {community_labels[cid]} — {len(comm)} nodes:")
    for lbl in labels:
        print(f"      · {lbl}")
    print()
```

## Step 8: VISUALIZATION 1 — Full Literature Network

**Reading the graph:**
- **Node size** = influence score (bigger = more central to literature)
- **Node color** = source type (empirical / report / policy / theory / YOUR STUDY)
- **Edge thickness** = number of shared themes (stronger link = thicker line)
- **Node position** = force-directed layout (closely related nodes cluster together)


```python
# ── VISUALIZATION 1: Full Network ───────────────────────────────────────────
fig, ax = plt.subplots(1, 1, figsize=(18, 13))
fig.patch.set_facecolor('#0d1117')
ax.set_facecolor('#0d1117')

# Layout
pos = nx.spring_layout(G, weight='weight', seed=42, k=2.8, iterations=100)

# Color map by type
type_colors = {
    "empirical": "#3b82f6",
    "report":    "#f59e0b",
    "policy":    "#8b5cf6",
    "theory":    "#10b981",
    "gap":       "#ef4444",
}

node_colors  = [type_colors[G.nodes[n]['ntype']] for n in G.nodes()]
node_sizes   = [G.nodes[n]['weight'] * 120 for n in G.nodes()]
edge_weights = [G[u][v]['weight'] for u, v in G.edges()]
edge_alphas  = [min(0.8, 0.15 + w * 0.18) for w in edge_weights]
edge_widths  = [0.4 + w * 0.5 for w in edge_weights]

# Draw edges
for i, (u, v) in enumerate(G.edges()):
    x0, y0 = pos[u]; x1, y1 = pos[v]
    ax.plot([x0, x1], [y0, y1],
            color='#4a5568', alpha=edge_alphas[i], linewidth=edge_widths[i], zorder=1)

# Draw nodes
nx.draw_networkx_nodes(G, pos, ax=ax,
    node_color=node_colors, node_size=node_sizes, alpha=0.92, linewidths=0.8,
    edgecolors=['#ffffff' if G.nodes[n]['ntype']=='gap' else '#1a1a2e' for n in G.nodes()])

# Labels
labels = {n: G.nodes[n]['label'] for n in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=labels, ax=ax,
    font_size=6.5, font_color='#e6edf3', font_family='serif')

# Legend
legend_elements = [
    mpatches.Patch(color="#3b82f6", label="Empirical Article"),
    mpatches.Patch(color="#f59e0b", label="Institutional Report"),
    mpatches.Patch(color="#8b5cf6", label="Policy Document"),
    mpatches.Patch(color="#10b981", label="Theoretical Framework"),
    mpatches.Patch(color="#ef4444", label="YOUR STUDY (gap)"),
    Line2D([0],[0], color='#4a5568', linewidth=1, label='1 shared theme'),
    Line2D([0],[0], color='#4a5568', linewidth=3, label='3+ shared themes'),
]
ax.legend(handles=legend_elements, loc='lower left', fontsize=9,
          facecolor='#161b22', edgecolor='#30363d', labelcolor='#e6edf3')

ax.set_title("CBC Teacher Preparedness — Literature Network (Kenya, 2026)\n"
             "Node size = influence score  |  Edge thickness = shared themes",
             fontsize=14, pad=16, color='#e6edf3', fontfamily='serif')
ax.axis('off')
plt.tight_layout()
plt.savefig('Visualizations/Full_network.png', dpi=180, bbox_inches='tight', facecolor='#0d1117')
plt.close()
print("✅ Figure 1 saved: fig1_full_network.png")
```
